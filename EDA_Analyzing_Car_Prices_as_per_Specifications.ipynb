# **Project Summary -**

This exploratory data analysis project investigates how various car specifications influence vehicle prices using an automobile dataset. The primary goal is to understand which technical and categorical attributes are most strongly associated with price so that future pricing and product decisions can be grounded in data rather than intuition alone. After basic cleaning, including handling missing values and inconsistent entries, the final working dataset contains 199 rows and 26 columns. The data‑wrangling stage focused on converting placeholder characters into meaningful numeric values or removing problematic records. After this visualizations reveal several important patterns in the data. The distribution of symboling (a risk rating related to insurance) shows multiple categories from −2 to +3 but, when average price is plotted by symboling, prices do not follow a clear increasing or decreasing trend with risk level. In contrast, analyses involving engine size, fuel economy, and price display stronger structure. Grouped bar charts show that average car price tends to increase with engine size, with higher engine‑size categories generally commanding higher mean prices than smaller engines, even though there is some noise between adjacent sizes. Fuel‑economy variables (city-mpg and highway-mpg) exhibit an inverse relationship with engine size: when mean engine size is plotted against combinations of city and highway mileage, higher mpg pairs are clearly associated with smaller engines, while low‑mpg combinations are linked to larger engines. A focused correlation heatmap on symboling, engine-size, and price quantifies these relationships. From a business perspective, the EDA provides concrete guidance. First, it emphasizes that engine specifications (size, and by extension horsepower) should feature prominently in pricing models and market positioning because they correlate strongly with what customers pay. Second, it shows that fuel‑efficiency metrics are tightly linked to engine size, supporting strategies such as offering clearly segmented “efficiency” and “performance” product lines with pricing that reflects both performance benefits and long‑term fuel‑cost implications. Third, it warns against over‑reliance on symboling as a pricing factor, suggesting that insurers may care about it more than retail buyers do. Overall, the analysis prepares a solid foundation for building predictive models of car prices and for refining pricing, assortment, and marketing strategies in the automotive domain.

##### 1. Why did you pick the specific chart?

Bar charts are designed to compare summary values (like means) between categories, making it easy to see which symboling levels are associated with higher or lower average prices. Symboling is a risk-rating category from about -3 (safer) to +3 (riskier), so plotting average price by each category shows how price changes with risk level in a clear, interpretable way.

##### 2. What is/are the insight(s) found from the chart?

After analyzing we found that average car price is not monotonic with the risk rating (symboling); some safer categories are as expensive as, or more expensive than, riskier ones. Symboling values around -2 and -1 (safer cars) and 3 (riskier cars) appear to have relatively high average prices, while mid‑range categories like 1 and 2 show lower means in this sample. This indicates only a weak or inconsistent relationship between insurance risk rating and price in the dataset.

##### 3. Will the gained insights help creating a positive business impact?
Are there any insights that lead to negative growth? Justify with specific reason.

Above data shows that average price does not change consistently with symboling (risk rating), suggesting symboling alone is a weak driver of price. This tells a business not to over‑weight this variable in pricing or product strategy and to focus instead on stronger drivers such as engine size, horsepower, or brand, which are known to influence car prices more strongly.

#### Chart - 2

````python
# Chart - 2 visualization code
df.groupby(['engine-size'])['city-mpg'].mean().plot(kind = 'bar')
````

````python
df.groupby(['engine-size'])['highway-mpg'].mean().plot(kind = 'bar')
````

````python
df.groupby(['highway-mpg','city-mpg'])['engine-size'].mean().plot(kind = 'line')
````

##### 1. Why did you pick the specific chart?

The x‑axis (highway-mpg, city-mpg) pairs are sorted from low to high mileage, so connecting the mean engine sizes with a line creates a visually continuous path that clearly emphasizes the general downward trend: higher mpg combinations correspond to smaller engines on average. This chart shows whether engine size typically rises or falls as fuel efficiency improves, a line chart quickly conveys this monotonic tendency in a single smooth profile, making the inverse relationship between engine size and fuel economy easy to spot at a glance.

##### 2. What is/are the insight(s) found from the chart?

Low‑mpg combinations on the left side of the chart correspond to much larger average engines, while high‑mpg combinations on the right are associated with noticeably smaller engines, confirming that bigger engines are generally less fuel efficient. Although there is some fluctuation point to point, the overall downward profile suggests that fuel‑economy improvements in this dataset are strongly linked to downsizing engines, which is consistent with real‑world vehicle design strategies to reduce fuel consumption.

##### 3. Will the gained insights help creating a positive business impact?
Are there any insights that lead to negative growth? Justify with specific reason.

The chart confirms a strong inverse relationship between engine size and fuel efficiency: vehicles with smaller engines achieve higher combined city–highway mpg, while larger engines are much less efficient. This insight helps manufacturers and product planners justify investment in engine downsizing, hybridization, or other efficiency technologies to meet fuel‑economy regulations and appeal to cost and eco‑conscious customers, which can boost sales and margins. ​​For pricing and marketing, the clear linkage between engine size and mpg supports segmenting offerings (e.g., “efficiency‑focused” small‑engine models vs “performance‑focused” large‑engine models) and setting prices that reflect both performance and lifetime fuel‑cost savings, improving value communication and potentially increasing conversion rates.

#### Chart - 14 - Correlation Heatmap

````python
# Correlation Heatmap visualization code
````

````python
df.head()
````

````python
df_corr = df[['symboling','engine-size','price']]
df_corr
````

````python
plt.figure(figsize=(10,3))
sns.heatmap(df_corr.corr(), cmap='coolwarm', annot=True)
````

##### 1. Why did you pick the specific chart?

This chart was used because a correlation heatmap is the most efficient way to show all pairwise correlations among several numerical variables in a single, easily readable view. The goal is to understand how symboling, engine-size, and price relate to one another; a correlation matrix with color encoding lets you see both the sign (positive/negative) and strength of each relationship at a glance.​

##### 2. What is/are the insight(s) found from the chart?

The chart shows that engine size has a strong positive linear relationship with price, while symboling has almost no linear relationship with either. The correlation between engine-size and price is about 0.87, which is considered strong and indicates that as engine size increases, price tends to increase as well. Symboling has weak negative correlations with both engine-size (around −0.11) and price (around −0.08), values that are close to zero and suggest symboling is not a useful linear predictor of either variable in this dataset.
